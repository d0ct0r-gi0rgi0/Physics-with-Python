### Exercise 2.1: Interpreting a Machine Number

Given the double-precision hexadecimal number:

$$\$4059000000000000$$

- Determine its **sign**, **mantissa**, and **exponent**.
- Express the corresponding value in **decimal notation**.


```python
# This function converts machine and hexadecimal numbers to binary.
def hex_to_bin(x):
    z=[]
    for k in range(0,len(str(x))):
        z.append(str(x)[k])
    x=z
    hexadecimal=['a','b','c','d','e','f']
    decimal=[10,11,12,13,14,15]

    for i in range(0,len(x)):
        for j in range(0,len(hexadecimal)):
            if x[i]==hexadecimal[j]:
                x[i]=str(decimal[j])
                
    for q in range(0,len(x)):
        x[q]=bin(int(x[q]))[2:]
        if len(x[q])==1:
            x[q]='000'+x[q]
        elif len(x[q])==2:
            x[q]='00'+x[q]
        elif len(x[q])==3:
            x[q]='0'+x[q]
    h=''
    for m in range(0,len(x)):
        h+=x[m]
    return h

hex_to_bin('4059000000000000')

```
