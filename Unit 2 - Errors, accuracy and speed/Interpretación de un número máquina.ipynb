### Ejercicio 2.1: interpretando un número máquina

dado el número en hexadecimal de precisión doble:

$$\$4059000000000000$$

- Calcular cual es el signo, la mantisa y el exponente. 
- Dar su número en decimal.

```python
# Esta función convierte números máquina y hexadecimales a binarios.
def hextobin(x):
    z=[]
    for k in range(0,len(str(x))):
        z.append(str(x)[k])
    x=z
    hexadecimal=['a','b','c','d','e','f']
    decimal=[10,11,12,13,14,15]

    for i in range(0,len(x)):
        for j in range(0,len(hexadecimal)):
            if x[i]==hexadecimal[j]:
                x[i]=str(decimal[j])
                
    for q in range(0,len(x)):
        x[q]=bin(int(x[q]))[2:]
        if len(x[q])==1:
            x[q]='000'+x[q]
        elif len(x[q])==2:
            x[q]='00'+x[q]
        elif len(x[q])==3:
            x[q]='0'+x[q]
    h=''
    for m in range(0,len(x)):
        h+=x[m]
    return h

hextobin('4059000000000000')
```

```python
# Esta función convierte números máquina y hexadecimales a decimales.
def hextodec(h):
    # Se pasa el número a binario:
    b=hextobin(str(h))
    print(len(b))
    # Primero sacamos el signo:
    s=int(b[0])
    
    # Ahora calculamos el exponente:
    bexp=11 # En precisión doble, el exponente tiene 11 bits
    exponente=int(b[1:bexp+1],2)-1023 # 1023 es el sesgo en precisión doble.
    print("El exponente es:",exponente)
    
    # Para la mantisa:
    bman=52 # tiene 52 bits
    man=b[12:]
    print(len(man))
    mantisa=1.0
    for i in range(0,bman):
        mantisa+=float(man[i])*2**(-i-1)
    print("La mantisa es:",mantisa)
    
    # Se hace el cálculo:
    N=((-1)**s)*mantisa*2**exponente
    print("El número decimal es:",N)
    return N

hextodec(4059000000000000)
```
