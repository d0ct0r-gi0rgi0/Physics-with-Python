## Exercise 5 – Fractality. The Mandelbrot Set

The Mandelbrot set, named in honor of its discoverer, the French mathematician **Benoît Mandelbrot**, is a fractal: an infinitely ramified mathematical object that contains structures within structures at arbitrarily fine scales.

The Mandelbrot set is defined using complex numbers as follows.

- Consider the equation:
  
  $$
  z' = z^2 + c,
  $$
  
  where \( z \) is a complex number and \( c \) is a complex constant. For any given value of \( c \), this equation produces a new complex number \( z' \) from an initial value \( z \).  
  The definition of the Mandelbrot set involves iterating this equation and includes all points in the complex plane that satisfy the following criterion:
  
  > *For a given value of \( c \), one starts with \( z = 0 \) and iterates the equation repeatedly.  
  > If at any iteration the condition \( |z'| > 2 \) is satisfied, then the point \( c \) does not belong to the Mandelbrot set. Otherwise, it does.*

- In principle, applying this definition would require iterating the equation infinitely many times in order to determine whether a given number belongs to the Mandelbrot set. However, in practice, it is sufficient to perform a finite number of iterations, typically around 100.

### Task

1. Write a program that generates an image of the Mandelbrot set for all complex values  
   \( c = x + i\,y \) such that
   \[
   -2 \leq x \leq 2, \qquad -2 \leq y \leq 2.
   \]
   Use a density plot in which points belonging to the Mandelbrot set are represented in black, and those that do not belong are represented in white.  
   The Mandelbrot set has a very characteristic shape, resembling a beetle-like figure with a protruding “snout”.

   *Hint:* Begin by analyzing a small number of points \( N \), and only increase the value of \( N \) to obtain the final result once the program is working correctly.

```python
def mandelbrot(N, max_iter):
    mandelbrot_set = np.zeros([N, N])
    interval = np.linspace(-2, 2, N)
    
    complex_matrix = np.zeros([N,N], complex)
    for i in range(N):
        for j in range(N):
            complex_matrix[i, j] = complex(interval[j], interval[i])

    for i in range(N):
            for j in range(N):
                iter = 0
                z0 = 0
                while abs(z0) <= 2 and iter <= max_iter:
                    iter += 1
                    z0 = z0**2 + complex_matrix[i, j]
                mandelbrot_set[i, j] = iter
    return mandelbrot_set

matrix = mandelbrot(1000, 100)
```

```python
plt.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Valor')
plt.title("Mandelbrot Set (c)")
plt.xlabel("Re(z)")
plt.ylabel("Im(x)")
plt.show()
```
