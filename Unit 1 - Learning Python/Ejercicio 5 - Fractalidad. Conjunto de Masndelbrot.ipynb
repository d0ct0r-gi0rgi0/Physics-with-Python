## Ejercicio 5 - Fractalidad. Conjunto de Mandelbrot.

El conjunto de Mandelbrot, nombrado en honor a su descubridor, el matemático francés Benôit Mandelbrot, es un fractal, un objeto matemático infinitamente ramificado que contiene estructuras dentro de estructura con cualquier profundidad con la que lo miremos. 

El conjunto de Mandelbrot se describe a partir de números complejos tal y como sigue.

* Sea la ecuación:
    $$z'=z^2+c,$$
  siendo $z$ un número complejo y $c$ una constante compleja. Para cualquier valor de $c$, esta ecuación proporciona un número $z'$ a partir de valor inicial $z$. La definición del conjunto de Mandelbrot involucra la iteracción de esta ecuación, y representa todos aquellos puntos del plano complejo que satisfacen la definición:
  
    "Para un valor dado de $c$, se empieza con $z=0$ y se itera repetidamente. Si el valor resultante cumple $|z'|> 2$ en algún momento, entonces el punto $c$ no pertenece al conjunto de Mandelbrot. De lo contrario si lo hace". 
    

* Para usar esta definición uno tendría, en principio, que iterar infinitamente la ecuación para probar si el número pertenece o no al conjunto de Mandelbrot.  Sin embargo, en la práctica es suficiente con iterar unas 100 veces.

1. Escribir un programa que obtenga una imagen del conjunto de Mandelbrot para todos aquellos valores complejos $c=x+i\,y$, tal que $-2\le x\le 2$ e $-2\le y\le 2$. Para ello utilizar un gráfico de densidad en el que los puntos que pertenezcan al conjunto se representen en negro y los que lo lo hagan en blanco. El conjunto de Mandelbrot es muy característico, se parece a una especie de escarabajo con una especie de hocico.  

    [Ayuda: empezar analizando un numéro pequeño de puntos $N$ y, sólo cuando el programa funcione correctamente, incrementar el valor de N para el resultado final].
---

```python
def mandelbrot(N, max_iter):
    mandelbrot_set = np.zeros([N, N])
    interval = np.linspace(-2, 2, N)
    
    complex_matrix = np.zeros([N,N], complex)
    for i in range(N):
        for j in range(N):
            complex_matrix[i, j] = complex(interval[j], interval[i])

    for i in range(N):
            for j in range(N):
                iter = 0
                z0 = 0
                while abs(z0) <= 2 and iter <= max_iter:
                    iter += 1
                    z0 = z0**2 + complex_matrix[i, j]
                mandelbrot_set[i, j] = iter
    return mandelbrot_set

matrix = mandelbrot(1000, 100)
```

```python
plt.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Valor')
plt.title("Conjunto de Mandelbrot (c)")
plt.xlabel("Re(z)")
plt.ylabel("Im(x)")
plt.show()
```
