## Ejercicio 3 - Probabilidad Binomial.
El coeficiente binomial es un entero definido por:

$$ \binom{n}{k} = \frac{n!}{k!\;(n-k)!}$$

para $k\ge1$ o $\binom{n}{0}=1$ para $k=0$.

---

1. Escribir una función en python que para cualquier n y k calcule el coeficiente binomial. Programarlo de la forma más eficiente posible. Asegurar que el resultado da un entero y que da el valor correcto para k=0. 

---

```python
def binomial_coeff(n, k):
    assert(type(n) == int)
    assert(type(k) == int)
    
    def factorial(N):
        assert(type(N) == int) 
        assert(N >= 0)
        if N == 0:
            fact = 1
        else:
            fact = 1
            for p in range(1, N+1):
                fact *= p
        return fact  
    
    if k == 0:
        result = 1
    else:
        result = factorial(n) / (factorial(k) * factorial(n-k))
    return int(result)

binomial_coeff(5, 2)
```

2. Usar este programa para representar por pantalla las primeras 20 líneas del triangulo de Pascal (la línea nth del triangulo contiene n+1 números, que son los coeficientes $\binom{n}{0}, \binom{n}{1}, \cdots,\binom{n}{n}$). 

---

```python
n = 20
lines = []
for k in range(0, n + 1):
    lines.append(binomial_coeff(n, k))
plt.plot(lines, color = 'mediumorchid')
plt.grid()
plt.title(f"{n} first lines of Pascal's Triangule")
plt.show()
```

3. La probabilidad de que lanzando una moneda $n$ veces se obtenga $k$ veces cara es $\binom{n}{k}/2^n$. Escribir una función que permita calcular la probabilidad de que lanzando una moneda $n$ veces se obtengan $k\le n$ veces cara. Obtener el resultado para n=100 y k=60. 

---

```python
def coin_flip(n, k):
    prob = binomial_coeff(n, k)/(2**n)
    return round(100*prob, 2)

coin_flip(100, 60)
```

4. Calcular la probabilidad de que lanzando 100 veces una moneda se obtengan más de 60 veces cara. 

---

```python
prob = 0
for k in range(61, 101):
    prob += coin_flip(100, k)
print(round(prob, 2))
```
